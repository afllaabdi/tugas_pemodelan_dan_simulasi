Class Agent

```python
import numpy as np
import matplotlib.pyplot as plt

class SleepAgent:
    def __init__(self):
        self.S = 0.7  # Sleep Quality awal
        self.R = 0.5  # Resilience
        self.P = 0.5  # Intervensi (default)
        self.history = []

    def update(self, D):
        # Model matematis
        self.S = self.S - D + (self.R * self.P)

        # Constraint
        self.S = max(0, min(1, self.S))

        # Simpan histori
        self.history.append(self.S)
```

Simulasi lingkungan

```python
def run_simulation(steps=100):
    agent = SleepAgent()

    for t in range(steps):
        # Lingkungan: stressor acak (sleep debt)
        D = np.random.uniform(0.1, 0.4)

        agent.update(D)

    return agent.history
```

Visualisasi

```python
result = run_simulation()

plt.plot(result)
plt.title("Simulasi Kualitas Tidur (Base Model)")
plt.xlabel("Time Step")
plt.ylabel("Sleep Quality")
plt.show()
```

Minggu 10

```python
def get_state(S):
    if S >= 0.7:
        return "Sehat"
    elif S >= 0.4:
        return "Terganggu"
    else:
        return "Insomnia"
```

```python
def run_simulation_uts(steps=100):
    agent = SleepAgent()
    states = []

    for t in range(steps):
        D = np.random.uniform(0.1, 0.4)
        agent.update(D)

        # ambil state tiap waktu
        states.append(get_state(agent.S))

    return agent.history, states
```

```python
result, states = run_simulation_uts()
```

```python
plt.figure(figsize=(10,5))
plt.plot(result)

# Tandai kondisi insomnia (opsional tapi keren)
for i, s in enumerate(states):
    if s == "Insomnia":
        plt.scatter(i, result[i])

plt.title("Simulasi Kualitas Tidur + State (UTS)")
plt.xlabel("Time Step")
plt.ylabel("Sleep Quality")
plt.show()
```
